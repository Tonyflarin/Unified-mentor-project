```python
def match_result(p):
    if p["Home Team Goals"]>p["Away Team Goals"]:
        return "Home Team Win"
    elif p["Home Team Goals"]<p["Away Team Goals"]:
        return "Home team loose"
    elif p["Home Team Goals"]==p["Away Team Goals"]:
        return "Draw_match"
```

```python
p["match_result"]=p.apply(lambda x: match_result(x),axis=1)
```

```python
p["match_result"].value_counts()
```
